```python
#Using a Heatmap to see the correlation between numerical variable using the palette OrRD.
sns.heatmap(data.corr(), annot=True, cmap="OrRd")
```

```python
#Vertical Bar plot to visualize the share between FeMale and Male:
g=sns.catplot(x="Gender", kind="count", palette="Set2", data=data)
#To include the value above the bar
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.015, 
            p.get_height() * 1.02, 
            '{0:.2f}'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
```

```python
#Vertical Bar Plot to visualize the proportion between Married and Single splited by the gender:
g=sns.catplot(x="Gender", kind="count", hue="Marital Status", palette="Set1", data=data)
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.015, 
            p.get_height() * 1.02, 
            '{0:.2f}'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
```

```python
#Vertical Bar Plot to visualize the share in the category Education:
from matplotlib import pyplot as plt
g=sns.catplot(x="Education", kind="count", palette="Set2", data=data)
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.015, 
            p.get_height() * 1.02, 
            '{0:.2f}'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
plt.xticks(rotation=45) #for a better visualization in the X axis
```

```python
#Vertical Bar Plot to visualize the share in the category Occupation:
=sns.catplot(x="Occupation", kind="count", palette="Set2", data=data)
ax = g.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.015, 
            p.get_height() * 1.02, 
            '{0:.2f}'.format(p.get_height()), 
            color='black', rotation='horizontal', size='large')
plt.xticks(rotation=45)
```

```python
#Using Scatter plot to see the distribution between Income and Age:
import numpy as np
sns.catplot(x="Income", y="Age", palette="Dark2", data=data, kind='swarm');
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#Just changing the format to box
sns.catplot(x="Income", y="Age", kind="box", data=data);
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#Returning to the previous format and including a differentiation about Marital Status:
sns.catplot(x="Income", y="Age", palette="Dark2", data=data, kind='swarm', hue="Marital Status");
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#Returning to the previous format and including a differentiation about Marital Status:
sns.catplot(x="Income", y="Age", palette="Dark2", data=data, kind='swarm', hue="Education", hue_order=['Partial High School','High School','Parcial College','Bachelors',"Graduate Degree"])
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#Using Scatter plot to see the distribution between Income and Age including a differentiation about Occupation 
#and put the values in order: Manual, Clerical, Skilled Manual, Professional and Management
sns.catplot(x="Income", y="Age", palette="Dark2", data=data, kind='swarm', hue="Occupation",hue_order=['Manual','Clerical','Skilled Manual','Professional',"Management"])
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#to allow the manipulation of the chart.From this point on every chart can be manipulated.
%matplotlib notebook 

#Using Scatter plot to see the distribution between Income and Age including a differentiation about Occupation 
#and put the values in order: Manual, Clerical, Skilled Manual, Professional and Management.
#Including diferent forms to separate the data between Married and Single.
sns.relplot(x="Income", y="Age", palette="Dark2", data=data, hue="Occupation",hue_order=['Manual','Clerical','Skilled Manual','Professional',"Management"],style="Marital Status" )
plt.yticks(np.arange(20,100,10))
plt.xticks(rotation=45)
```

```python
#Trying to find a correlation between Income, Number of cars and bike buyer or not. Including a tendency line between Income and number of cars:
sns.lmplot(x="Income",y="Cars", palette="Set2", data=data, hue="Purchased Bike");
```

```python
#From this point on charts can't be manipulated.
%matplotlib inline

#Using a Pair Plot to visualize correlation and distribution between the numerical categories:
sns.pairplot(data,hue="Purchased Bike", diag_kind="hist",palette="Set1");
```
